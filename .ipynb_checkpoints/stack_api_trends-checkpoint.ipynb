```python
from stackapi import StackAPI
from datetime import datetime, timedelta
import pandas as pd
```

```python
SITE = StackAPI('stackoverflow')
```

```python
delta = timedelta(days=365)
start_date = datetime(year=2009, month=1, day=1)
end_date = start_date + delta
```

```python
type(end_date.year)
```

```python
print(start_date.strftime("%d %B, %Y")," to ", end_date.strftime("%d %B, %Y"))
```

```python
questions = SITE.fetch('questions', fromdate=start_date,todate=end_date, tagged='python', sort='votes')
```

```python
total_views = 0
for question in questions['items']:
    total_views += question["view_count"]
```
